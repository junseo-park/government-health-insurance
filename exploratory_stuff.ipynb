```python
# Making a new dataset suitable for graphing purposes
bar_life = life.fillna(0)
bar_life.isnull().sum()
```

```python
bar_life.head(20)
```

```python
# changing attributes from percent of gdp to actual $$$$
# net aid excluded because there are negative values

attributes = ["health", "education", "military", "research"]

def percent_to_dollar(row, column):
    dollar = (row[column]/100)*row["gdp"]
    return dollar

for i in attributes:
    bar_life[i] = bar_life.apply(percent_to_dollar, column=i, axis=1)
```

```python
# Testing on small subset of USA

trace1 = Bar(
    x=bar_life[bar_life["country"]=="United States"]["year"],
    y=bar_life[bar_life["country"]=="United States"]["gdp"],
    name='gdp'
)
trace2 = Bar(
    x=bar_life[bar_life["country"]=="United States"]["year"],
    y=bar_life[bar_life["country"]=="United States"]["health"],
    name='health'
)
trace3 = Scatter(
    x=bar_life[bar_life["country"]=="United States"]["year"],
    y=bar_life[bar_life["country"]=="United States"]["life_expectancy"],
    name='life expectancy',
    yaxis='y2'
)
data = [trace1, trace2, trace3]
layout = Layout(
    title='GDP',
    barmode="group",
    yaxis=dict(
        title='GDP'
    ),
    yaxis2=dict(
        title='Life Expectancy',
        titlefont=dict(
            color='rgb(148, 103, 189)'
        ),
        tickfont=dict(
            color='rgb(148, 103, 189)'
        ),
        overlaying='y',
        side='right'
    )
)
fig = Figure(data=data, layout=layout)
py.iplot(fig, filename='multiple-axes-grouped-bar')
```

```python
# Setup for final graph 

list_of_traces = []
countries = bar_life["country"].unique()
attributes = ["gdp", "health", "education", "military", "research"]

for country in countries:
    for a in attributes:
        list_of_traces.append(Bar(
    x=bar_life[bar_life["country"]==country]["year"],
    y=bar_life[bar_life["country"]==country][a],
    name=a,
    visible=False
))
    list_of_traces.append(Scatter(
    x=bar_life[bar_life["country"]==country]["year"],
    y=bar_life[bar_life["country"]==country]["life_expectancy"],
    name='life expectancy',
    yaxis='y2',
    visible=False
))

    
list_of_buttons = []
for index, country in enumerate(countries):
    list_of_buttons.append(dict(args=["visible", 
                                      [true_false_button(i,index,num_att=len(attributes)) for i in range(0,(len(attributes)+1)*len(countries))]],
                                label=country,
                                method="restyle"))
    


                          
```
