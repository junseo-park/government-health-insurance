```python
def delete_rows(df, rows_to_delete):
    for i in range(len(df)):
        if df.loc[i, 'Country Name'] in rows_to_delete:
            df.drop(i, inplace=True)

educ_exp = pd.read_csv('data/raw_data/education_expenditure/API_SE.XPD.TOTL.GD.ZS_DS2_en_csv_v2.csv', header=2)
delete_rows(educ_exp, to_remove)

educ_exp.head()
```

```python
import os

def clean_csv(dir_name):
    RAW_DATA_PATH = os.path.join('data', 'raw_data', dir_name)
    filenames = os.listdir(RAW_DATA_PATH)
    RAW_FILE_PATH = filenames[0]
    
    df = pd.read_csv(os.path.join(RAW_DATA_PATH, RAW_FILE_PATH), header=2)
    delete_rows(df, to_remove)
    df.drop(['Country Code', 'Indicator Name', 'Indicator Code', '2015', '2016', 'Unnamed: 61'], axis=1, inplace=True)
    df.drop([str(x) for x in range(1960, 1995)], axis=1, inplace=True)
    df.rename(columns={'Country Name':'country_name'}, inplace=True)
    
    CLEAN_DATA_PATH = os.path.join('data', 'clean_data', dir_name, )
    df.to_csv(CLEAN_DATA_PATH + '.csv', index=False)
    
dir_names = ['education_expenditure', 'gdp', 'health_expenditure', 'infant_mortality', 
             'life_expectancy', 'military_expenditure', 'net_aid']

for dir_name in dir_names:
    if not os.path.exists('data/clean_data/' + dir_name + '.csv'):
        clean_csv(dir_name)
    else:
        print(dir_name + '.csv already exists.')
```

```python
def read_my_csv(filename):
    DATA_PATH = os.path.join('data', 'clean_data', filename)
    return pd.read_csv(DATA_PATH, index_col='country_name')

educ_exp = read_my_csv('education_expenditure.csv')
gdp = read_my_csv('gdp.csv')
health_exp = read_my_csv('health_expenditure.csv')
inf_mort = read_my_csv('infant_mortality.csv')
life_exp = read_my_csv('life_expectancy.csv')
mili_exp = read_my_csv('military_expenditure.csv')
net_aid = read_my_csv('net_aid.csv')

all_df = [educ_exp, gdp, health_exp, inf_mort, life_exp, mili_exp, net_aid]
all_df_names = ['educ_exp', 'gdp', 'health_exp', 'inf_mort', 'life_exp', 'mili_exp', 'net_aid']
```

```python
for i, val_name in zip(range(len(all_df)), all_df_names):
    df = all_df[i]
    df.reset_index(inplace=True)
    df = df.melt(id_vars='country_name', var_name='year', value_name=val_name)
    df['country_name_year'] = df['country_name'].map(str) + df['year'].map(str)
    df.drop(['country_name', 'year'], axis=1, inplace=True)
    df.set_index('country_name_year', inplace=True)
    all_df[i] = df
```

```python
master_df = all_df[0].join([x for x in all_df[1:]])
master_df = master_df[['educ_exp', 'gdp', 'health_exp', 'mili_exp', 'inf_mort', 'life_exp']]
```

# Linear Regression

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression

X = master_df.loc[:, 'educ_exp':'mili_exp']
y = master_df.loc[:, 'life_exp']
X_train, X_test, y_train, y_test = train_test_split(X, y)

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
print('cross_val_score:', cross_val_score(lin_reg, X_train, y_train))

y_pred_lin_reg = lin_reg.predict(X_test)
print('rmse:', np.sqrt(mean_squared_error(y_test, y_pred_lin_reg)))
```

# Single Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(max_depth=3, random_state=42)
tree_reg.fit(X_train, y_train)
print('cross_val_score:', cross_val_score(tree_reg, X_train, y_train))

y_pred_tree_reg = tree_reg.predict(X_test)
print('rmse:', np.sqrt(mean_squared_error(y_test, y_pred_tree_reg)))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'max_leaf_nodes': list(range(2, 100)), 'min_samples_split': list(range(2, 11))}
grid_search_decision_tree = GridSearchCV(DecisionTreeRegressor(random_state=42), param_grid, n_jobs=-1)
grid_search_decision_tree.fit(X_train, y_train)

print('best params:', grid_search_decision_tree.best_params_)

print('cross_val_score:', cross_val_score(grid_search_decision_tree.best_estimator_, X_train, y_train))

y_pred_tree_grid = grid_search_decision_tree.best_estimator_.predict(X_test)
print('rmse:', np.sqrt(mean_squared_error(y_test, y_pred_tree_grid)))
```

# Random Forest

```python
from sklearn.ensemble import GradientBoostingRegressor

gbrt = GradientBoostingRegressor(max_depth=3, n_estimators=3, learning_rate=1.0, random_state=42)
gbrt.fit(X_train, y_train)
print('cross_val_score:', cross_val_score(gbrt, X_train, y_train))
y_pred_gbrt = gbrt.predict(X_test)
print('rmse:', np.sqrt(mean_squared_error(y_test, y_pred_gbrt)))
```

```python
# Early Stopping

X_train_2, X_val, y_train_2, y_val = train_test_split(X_train, y_train, random_state=42)

gbrt = GradientBoostingRegressor(max_depth=2, n_estimators=1000, random_state=42)
gbrt.fit(X_train_2, y_train_2)

errors = [mean_squared_error(y_val, y_pred)
          for y_pred in gbrt.staged_predict(X_val)]
bst_n_estimators = np.argmin(errors)

gbrt_best = GradientBoostingRegressor(max_depth=2,n_estimators=bst_n_estimators, random_state=42)
print('Early stopping found', gbrt_best.n_estimators)
gbrt_best.fit(X_train, y_train)

print('cross_val_score:', cross_val_score(gbrt_best, X_train, y_train))
y_pred_gbrt_best = gbrt.predict(X_test)
print('rmse:', np.sqrt(mean_squared_error(y_test, y_pred_gbrt_best)))
```

```python
from scipy.stats import reciprocal

param_grid = {'learning_rate':[0.01, 0.05, 0.1, 0.5, 1.0],
             'max_depth':np.arange(2, 5),
             'min_samples_split':[2, 3],
              'loss':['ls', 'huber']}

gbrt_grid = GridSearchCV(GradientBoostingRegressor(n_estimators=bst_n_estimators, random_state=42),
                        param_grid=param_grid, n_jobs=-1, verbose=1)
gbrt_grid.fit(X_train, y_train)
gbrt_grid.best_params_
```

```python
print('cross_val_score:', cross_val_score(gbrt_grid.best_estimator_, X_train, y_train))
y_pred_gbrt_grid_best = gbrt_grid.best_estimator_.predict(X_test)
print('rmse:', np.sqrt(mean_squared_error(y_test, y_pred_gbrt_grid_best)))
```

# Support Vector Machines

```python
from sklearn.svm import LinearSVR
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

lin_svr = LinearSVR(random_state=42)
lin_svr.fit(X_train_scaled, y_train)
print('cross_val_score:', cross_val_score(lin_svr, X_train_scaled, y_train))
y_pred_lin_svr = lin_svr.predict(X_test_scaled)
print('rmse:', np.sqrt(mean_squared_error(y_test, y_pred_lin_svr)))
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import reciprocal, uniform

param_distributions = {'kernel':['poly', 'sigmoid', 'rbf'],
                       "gamma": reciprocal(0.001, 0.1), 
                       "C": uniform(1, 10)}
rnd_search_cv = RandomizedSearchCV(SVR(), param_distributions, n_iter=25, random_state=42)
rnd_search_cv.fit(X_train_scaled, y_train)

print('cross_val_score:', cross_val_score(rnd_search_cv.best_estimator_, X_train_scaled, y_train))
y_pred_rnd_cv = rnd_search_cv.best_estimator_.predict(X_test_scaled)
print('rmse:', np.sqrt(mean_squared_error(y_test, y_pred_rnd_cv)))
```

# Neural Networks

```python
import tensorflow as tf
from keras.models import Model, Sequential
from keras.layers import Dense, Activation, Flatten, Dropout
from keras.wrappers.scikit_learn import KerasRegressor

np.random.seed(42)

def base_model():
    model = Sequential()
    model.add(Dense(4, input_dim=4, init='normal', activation='relu'))
    model.add(Dense(1, init='normal'))
    
    model.compile(optimizer='adam',loss='mean_squared_error')
    return model

estimator = KerasRegressor(build_fn=base_model, nb_epoch=100, batch_size=5)
results = cross_val_score(estimator, X_train_scaled, y_train)
print("Results: %.2f (%.2f) MSE" % (results.mean(), results.std()))
#estimator.fit(X_train_scaled, y_train)

#estimator.score(y_test, estimator.predict(X_test_scaled))
```

```python
from sklearn.model_selection import KFold
from sklearn.pipeline import Pipeline

results = cross_val_score(estimator, X_train_scaled, y_train)
print("Results: %.2f (%.2f) MSE" % (results.mean(), results.std()))

print('cross_val_score:', results)
y_pred_base_nn = estimator.predict(X_test_scaled)
print('rmse:', np.sqrt(mean_squared_error(y_test, y_pred_base_nn)))
```

```python
def deeper_model():
    model = Sequential()
    model.add(Dense(4, input_dim=4, kernel_initializer='normal', activation='relu'))
    model.add(Dense(4, kernel_initializer='RandomNormal', activation='relu'))
    model.add(Dense(2, kernel_initializer='normal'))
    model.add(Dense(1, kernel_initializer='normal'))
    
    model.compile(optimizer='adam',loss='mean_squared_logarithmic_error')
    return model

pipeline = Pipeline([
    ('std_scl', StandardScaler()),
    ('keras_regress', KerasRegressor(build_fn=base_model, nb_epoch=100, batch_size=5))
])

kfold = KFold(n_splits=10, random_state=42)
results = cross_val_score(pipeline, X_train, y_train, cv=kfold)
print("Results: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```
